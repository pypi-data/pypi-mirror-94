```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def make_data(N, f=0.3, rseed=1):
    rand = np.random.RandomState(rseed)
    x = rand.randn(N)
    x[int(f * N):] += 5
    return x

x = make_data(20)
x = [14, 15.5, 29]
y = [1.5, 13.5, 13.5]
x_d = np.linspace(-0.5, 30.5, 1000)
# density = sum(norm(xi).pdf(x_d) for xi in x)
density = sum(norm(xi, yi).pdf(x_d) for xi, yi in zip(x,y))

plt.fill_between(x_d, density, alpha=0.5)
plt.plot(x, np.full_like(x, -0.01), '|k', markeredgewidth=1)

plt.axis([-0.5, 30.5, -0.02, 0.33]);
```

```python
from optuna import study
from optuna.samplers import tpe
from sklearn import svm, ensemble, datasets, metrics, model_selection

X, y = datasets.load_boston(return_X_y=True)
X_train, X_val, y_train, y_val = model_selection.train_test_split(X, y, random_state=0)

DOMAIN = []

for alg in ["RF", "GBRT", "ET"]:
    for i in range(2,33):
        DOMAIN.append((("alg", alg), ("depth", i)))

# Define an objective function to be minimized.
def objective(trial):

    # Invoke suggest methods of a Trial object to generate hyperparameters.
    regressor_name = trial.suggest_categorical("alg", ['RF', 'GBRT', "ET"])
    max_depth = trial.suggest_int('depth', 2, 32)
     
    if regressor_name == 'RF':
        regressor_obj = ensemble.GradientBoostingRegressor(max_depth=max_depth)
    elif regressor_name == "GBRT":
        regressor_obj = ensemble.RandomForestRegressor(max_depth=max_depth)
    elif regressor_name == "ET":
        regressor_obj = ensemble.ExtraTreesRegressor(max_depth=max_depth)
    
    trial.setter()

        
#     if trial.check():
#         print("hi", trial.params)
#         cstrs = trial.get_new_random_params([5, 5], "int")
    print("hi", trial.params)

    regressor_obj.fit(X_train, y_train)
    y_pred = regressor_obj.predict(X_val)

    error = metrics.mean_squared_error(y_val, y_pred)

    return error  # An objective value linked with the Trial object.


study = study.create_study(sampler=tpe.TPESampler(), domain=DOMAIN) 
study.optimize(objective, n_trials=20)  # Invoke optimization of the objective function
# print(study.best_params)
```
